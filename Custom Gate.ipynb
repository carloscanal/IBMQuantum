<h1>Create a custom BellState gate in Qiskit</h1>

```python
from qiskit import QuantumCircuit

# Create a new circuit with the custom gate
bellState = QuantumCircuit(2,name="Bell State")
 
# Add a Hadamard gate to qubit 0
bellState.h(0)
 
# Perform a controlled-X gate on qubit 1, controlled by qubit 0
bellState.cx(0, 1)

bellState.draw("mpl")

# Now, we create a new circuit with two copies of the custom gate

qc = QuantumCircuit(4)

qc.append(bellState,[0,1])
qc.append(bellState,[2,3])

qc.draw("mpl")
```
